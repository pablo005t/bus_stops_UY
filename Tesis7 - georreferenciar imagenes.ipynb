```python
# --- Parámetros ---
STEP_METERS = 50  # 40–60 m da buen solapamiento a zoom 20
ZOOM = 20
SIZE = (640, 640)
SCALE = 1
MAPTYPE = "satellite"
OUTPUT_DIR = "/content/drive/My Drive/Tesis/Datos/Imagenes_Satelitales_Ruta_8"

import os, time, random, numpy as np, geopandas as gpd
from shapely.geometry import LineString, MultiLineString
from shapely.ops import unary_union
from pyproj import Transformer

os.makedirs(OUTPUT_DIR, exist_ok=True)

# 1) Asegurar WGS84 (lat/lon)
route_8 = route_8.to_crs(4326)

# 2) Unir toda la geometría de la Ruta 8
merged = unary_union(route_8.geometry)

# 3) Pasar a metros (EPSG:3857), muestrear cada STEP_METERS, volver a 4326
g3857 = gpd.GeoSeries([merged], crs=4326).to_crs(3857).iloc[0]

def iter_lines(g):
    if isinstance(g, LineString):
        yield g
    elif isinstance(g, MultiLineString):
        for gg in g.geoms:
            # En algunos datasets puede anidar más niveles
            if isinstance(gg, (LineString, MultiLineString)):
                yield from iter_lines(gg)

# Muestreo ordenado con deduplicación estable
pts_3857 = []
for line in iter_lines(g3857):
    L = float(line.length)
    # desde 0 hasta L, paso STEP_METERS, e incluir extremo final
    dists = np.arange(0, max(L, 1.0), STEP_METERS, dtype=float).tolist()
    if not dists or dists[-1] < L:
        dists.append(L)
    for d in dists:
        pts_3857.append(line.interpolate(d))

to4326 = Transformer.from_crs(3857, 4326, always_xy=True)

coords = []
seen = set()
for p in pts_3857:
    lat = to4326.transform(p.x, p.y)[1]
    lon = to4326.transform(p.x, p.y)[0]
    key = (round(lat, 6), round(lon, 6))
    if key not in seen:
        seen.add(key)
        coords.append((lat, lon))

print(f"Puntos a descargar: {len(coords)} (paso ~{STEP_METERS} m)")

# 4) Descarga con tu función test_static_maps_key
def outname(i):
    return os.path.join(OUTPUT_DIR, f"ruta8_pt_{i:05d}_z{ZOOM}.png")

remaining = [(i, lat, lon) for i, (lat, lon) in enumerate(coords, start=1)
             if not os.path.exists(outname(i))]
"""
print(f"Total puntos: {len(coords)} | Faltan: {len(remaining)}")

errores = []
for i, lat, lon in remaining:
    try:
        img = test_static_maps_key(
            lat, lon,
            zoom=ZOOM, size=SIZE, scale=SCALE, maptype=MAPTYPE
        )
        with open(outname(i), "wb") as f:
            f.write(img)
        if i % 100 == 0:
            print(f"✅ Guardadas hasta el índice {i}")
    except Exception as e:
        msg = str(e)[:200]
        errores.append((i, lat, lon, msg))
        print(f"❌ {i} ({lat:.6f},{lon:.6f}) -> {msg}")
    # ~1–2 req/s con jitter para evitar 429
    time.sleep(0.25 + random.random() * 0.25)

print("Hecho. Carpeta:", OUTPUT_DIR)
if errores:
    print(f"Quedaron {len(errores)} errores (podes reintentar solo esos).")"""
```

```python
# ====== 1) Montar Drive ======
from google.colab import drive
drive.mount('/content/drive')

# ====== 2) Configuración (EDITA ESTO) ======
IMG_FOLDER = "/content/drive/MyDrive/Tesis/Datos/Ruta8/paradas_reales"  # carpeta con .png
CSV_PATH   = "/content/drive/MyDrive/Tesis/Datos/remaining_route8_points.csv"  # csv con ['index','latitude','longitude']
OUTPUT_FOLDER = "/content/drive/MyDrive/Tesis/Datos"

# ====== 3) Imports ======
import os, glob, re
import pandas as pd
import numpy as np

# ====== 4) Cargar lista de imágenes (.png) ======
img_paths = glob.glob(os.path.join(IMG_FOLDER, '**', '*.png'), recursive=True)
df_imgs = pd.DataFrame({'image_path': img_paths})
df_imgs['image_name'] = df_imgs['image_path'].apply(os.path.basename)

print(f"Imágenes .png encontradas: {len(df_imgs):,}")

# ====== 5) Extraer índice desde el nombre (exacto 5 dígitos) ======
IDX_REGEX_5 = re.compile(r'(?<!\d)(\d{5})(?!\d)')   # bloque de 5 dígitos aislados

def extract_index(name: str):
    m = IDX_REGEX_5.search(name)
    if m:
        return int(m.group(1))  # "00023" -> 23
    return np.nan

df_imgs['index'] = df_imgs['image_name'].apply(extract_index)
missing_idx = df_imgs['index'].isna().sum()
if missing_idx:
    print(f"ADVERTENCIA: {missing_idx} archivo(s) no tienen un bloque de 5 dígitos en el nombre.")

# ====== 6) Leer CSV (columnas fijas) ======
df_meta = pd.read_csv(CSV_PATH)
expected_cols = ['index', 'latitude', 'longitude']
if list(df_meta.columns) != expected_cols and not set(expected_cols).issubset(df_meta.columns):
    raise ValueError(f"El CSV debe tener columnas exactamente {expected_cols}. "
                     f"Columnas encontradas: {list(df_meta.columns)}")

# nos quedamos solo con esas columnas y tipificamos
df_meta = df_meta[expected_cols].copy()
df_meta['index'] = pd.to_numeric(df_meta['index'], errors='coerce').astype('Int64')
df_meta = df_meta.rename(columns={'latitude':'lat', 'longitude':'lon'})

# ====== 7) Chequeos rápidos antes del merge ======
dup_imgs = df_imgs['index'].duplicated(keep=False).sum()
dup_meta = df_meta['index'].duplicated(keep=False).sum()

if dup_imgs:
    print(f"ATENCIÓN: hay {dup_imgs} filas de imágenes con índices repetidos (variass .png para el mismo índice).")
if dup_meta:
    print(f"ATENCIÓN: hay {dup_meta} filas en el CSV con índices repetidos.")

# ====== 8) Merge ======
merged = df_imgs.merge(df_meta, on='index', how='left')

sin_match = merged['lat'].isna().sum()
print(f"Merge listo. Filas: {len(merged):,}.  Sin coordenadas (no match en CSV): {sin_match:,}")

# ====== 9) Guardar resultados ======
OUT_CSV = os.path.join(OUTPUT_FOLDER, "paradas_reales_con_coords.csv")
merged.to_csv(OUT_CSV, index=False)
print(f"✔ Archivo guardado: {OUT_CSV}")
```
